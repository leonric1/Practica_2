```python
'''
Práctica 2 | Simulación De Cola Del IMSS

Pregunta 1

Equipo : Magno Gomez Ricardo Leon
         Ramirez Martinez Maya
'''
# Bibliotecas

import numpy as np
import random

# Variables

reloj = 0
tiempo_atencion = 30
numero_pacientes = 50000

# Clases

class Cola:
  """Representa una Cola del IMSS en la que los diferentes tipos de pacientes
  esperaran a ser atendidos"""
  def __init__(self): # Construcción de la cola
    """Crea una cola en la que pacientes
    del IMSS esperaran"""
    self.cola = []

  def __str__(self): # Para que este escrita bonita la cola
    """Regresa los elementos de la cola en forma de cadena"""
    return str(self.cola)

  def formar(self, paciente): # cola.formar(x)
    """Forma en la cola a un paciente"""
    self.cola.append(paciente)

  def desformar(self): # cola.desformar()
    """Desforma de la cola a un paciente"""
    if not self.cola:
      return None
    return self.cola.pop(0)

  def top(self): # cola.top()
    """Muestra a el ultimo paciente en la cola"""
    if len(self.cola) == 0:
      return None
    return self.cola[-1]

  def esta_vacia(self): # cola.esta_vacia()
    """Menciona la ausencia o existencia de pacientes en la cola"""
    if len(self.cola) == 0:
     print("La cola está vacía")
    else:
      print("La cola no está vacía, el último elemento es:", self.cola.top())

class Paciente:
  """Clase que representa a cada uno de los pacientes que entraran en la simulacion """
  def __init__(self, numero_de_paciente, hora_de_llegada, tipo_de_paciente):
    """Constructor"""
    self.numero_de_paciente = numero_de_paciente
    self.hora_de_llegada = hora_de_llegada
    self.tipo_de_paciente = tipo_de_paciente

class Simulacion:
  """Clase, representa el sistema que estamos simulando del IMSS"""
  def __init__(self, reloj, tiempo_atencion, numero_pacientes):
    """Esta funcion es nuestro constructor de la clase simulacion"""
    self.reloj = reloj
    self.tiempo_atencion = tiempo_atencion
    self.numero_pacientes = numero_pacientes

# Funciones

def asignación_tipo_paciente(numero_random):
  """Asigna el tipo de paciente, urgencias o rutinario según un numero
  aleatorio"""
  if numero_random < 0.2:
    return "urgencias"
  else:
    return "rutinario"

# CUERPO MAIN

def practica_2():
  """Funcion encargada de ejecutar la simulacion de la clinica"""
  contador_tiempos_totales = 0
  contador_tiempos_de_espera = 0

  print("| + + + + + + + + + + |  Clinica IMSS  | + + + + + + + + + + |")
  print("")
  reloj = 0

  for paciente in range(1, numero_pacientes + 1):
    tipo_de_paciente = asignación_tipo_paciente(np.random.random())
    tiempo_de_interarribo = round(random.uniform(0.5, 50.0), 2)
    hora_de_llegada = round(reloj + tiempo_de_interarribo,2)
    reloj += tiempo_de_interarribo

    paciente_atendido = Paciente(paciente, hora_de_llegada, tipo_de_paciente)

    if paciente_atendido.numero_de_paciente == 1:
      tiempo_esperando = 0

      hora_de_salida = round((hora_de_llegada + tiempo_esperando +
                              tiempo_atencion),2)

    else:
      tiempo_esperando = round((hora_de_salida - hora_de_llegada),2)

      if tiempo_esperando < 0:
        tiempo_esperando = 0

      hora_de_salida = round((hora_de_llegada + tiempo_esperando +
                              tiempo_atencion),2)

    contador_tiempos_totales += hora_de_salida - hora_de_llegada
    contador_tiempos_de_espera += tiempo_esperando

  promedio_tiempo_en_el_sistema = contador_tiempos_totales / numero_pacientes
  promedio_tiempo_de_espera = contador_tiempos_de_espera / numero_pacientes

  print("Tiempo de espera promedio: " +
        str(round(promedio_tiempo_de_espera, 2)))

  print("Tiempo promedio en el sistema: " +
        str(round(promedio_tiempo_en_el_sistema, 2)))

if __name__ == '__main__':
  practica_2()
```
